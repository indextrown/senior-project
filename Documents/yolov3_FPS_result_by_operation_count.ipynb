# 연산량 별 다중 영상처리 객체검출 성능(FPS) 결과

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### YOLO 연산량 별 FPS 결과 데이터

```python
#데이터
data = {
    "FP32": [9.80, 4.87, 3.0, 2.0],
    "FP16": [33.61, 17.81, 12.13, 8.94],
    "Int8": [59.61, 34.28, 20.94, 17.94] 
}
```

```python
#x축 인덱스
index = [1,2,3,4]

# 데이터프레임 생성
df = pd.DataFrame(data).T
```

### 연산량 별 결과 비교

```python
# 그래프 그리기
ax = df.T.plot(marker="o", figsize=(10, 6))
plt.xlabel("Operation Count")
plt.xticks(range(0,4), index)
plt.ylabel("FPS", rotation = 0)
plt.title("Yolov3 FPS Results by Operation Count")
plt.legend(title="Data Type", loc="upper right")

for idx, col_name in enumerate(df.columns):
    for i, val in enumerate(df[col_name]):
        ax.text(idx, val + 0.9, f'{val:.2f}', ha='center', va='bottom')

plt.show()
```

### 결과: 
#### 연산량이 낮아질수록 객체검출 추론속도가 빨라짐.
#### 영상 개수가 늘어날 수록 엣지 디바이스 성능 저하.
